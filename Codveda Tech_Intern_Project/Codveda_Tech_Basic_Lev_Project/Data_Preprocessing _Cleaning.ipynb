```python
# #After scraping, the next important step is data type cleaning and transformation.
# Right now:

# Price is stored as text (string)

# Rating is also text

# Analysis and visualization require numeric values

# fixing that and also converting the price from Pounds (£) to Nigerian Naira (₦).


# Check the data type of each column
# This helps us identify which columns need conversion

df.dtypes
```

```python

# -------------------------------
# STEP 1: Clean and convert Price (£)
# -------------------------------

# 1️⃣ Remove any non-numeric characters (like 'Â' or '£') using regex
# 2️⃣ Convert the cleaned string to float so we can do numeric operations
df["Price (£)"] = df["Price (£)"].replace(r'[^0-9.]', '', regex=True).astype(float)

# 3️⃣ Optional: convert to Naira using a fixed exchange rate
exchange_rate = 2000  # £1 ≈ ₦2000
df["Price (₦)"] = df["Price (£)"] * exchange_rate

# -------------------------------
# STEP 2: Convert Rating text to numeric
# -------------------------------

# 1️⃣ Strip whitespace from Rating column (in case of extra spaces)
df["Rating"] = df["Rating"].astype(str).str.strip()

# 2️⃣ Map textual ratings to numeric values
# "One" → 1, "Two" → 2, ..., "Five" → 5
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

df["Rating"] = df["Rating"].map(rating_map)

# 3️⃣ Handle any missing ratings (e.g., if some books had no rating)
# Fill missing values with 0 to indicate "no rating"
df["Rating"] = df["Rating"].fillna(0).astype(int)

# -------------------------------
# STEP 3: Quick check
# -------------------------------
print(df.head(10))
```

```python
# Ensure correct types
df["Price (£)"] = df["Price (£)"].astype(float)
df["Price (₦)"] = df["Price (₦)"].astype(float)
df["Rating"] = df["Rating"].astype(int)
df["Title"] = df["Title"].astype(str)
df["Availability"] = df["Availability"].astype(str)
df
```

```python
# STEP: Categorize books by Price in Naira (₦)
# -------------------------------

# Define bin edges for Naira
# Multiply the previous pound bins by exchange rate (₦2000)
bins_naira = [0, 40000, 80000, 120000, 200000, 400000]  

# Labels for each price range
labels = ["Very Low", "Low", "Medium", "High", "Very High"]

# Create a new column 'Price Range (₦)'
df["Price Range (₦)"] = pd.cut(df["Price (₦)"], bins=bins_naira, labels=labels)

# -------------------------------
# Check the result
# -------------------------------
print(df[["Title", "Price (₦)", "Price Range (₦)"]].head(10))
```

```python
# Categorize rating 
df["Rating Group"] = pd.cut(df["Rating"], bins=[0,2,4,5], labels=["Low","Medium","High"])
df
```

```python
# Print clean Data
df.to_csv("books_cleaned.csv", index=False)
```
